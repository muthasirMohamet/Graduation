```python
import asyncio
import sys
from playwright.async_api import async_playwright
import json

# On Windows, set the Proactor event loop policy.
if sys.platform.startswith('win'):
    asyncio.set_event_loop_policy(asyncio.WindowsProactorEventLoopPolicy())

async def scrape_tiktok_video(video_url):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        await page.goto(video_url)
        await page.wait_for_timeout(3000)  # Wait for the page to load

        # Extract video details (update selectors as needed)
        video_info = {
            "url": video_url,
            "likes": await page.text_content('div[data-e2e="digg-count"]'),
            "comments": await page.text_content('div[data-e2e="comment-count"]'),
            "shares": await page.text_content('div[data-e2e="share-count"]'),
            "views": await page.text_content('div[data-e2e="play-count"]'),
            "author": await page.text_content('h3[class*="username"]'),
            "description": await page.text_content('h1[data-e2e="video-desc"]')
        }
        print("Video Details:")
        print(json.dumps(video_info, indent=4))

        # Extract comments
        comment_elements = await page.query_selector_all('div[data-e2e="comment-item"]')
        comment_list = []
        for comment in comment_elements:
            try:
                username = await comment.query_selector('span[class*="user-name"]').text_content()
            except Exception:
                username = "N/A"
            try:
                text = await comment.query_selector('div[class*="text"]').text_content()
            except Exception:
                text = ""
            try:
                likes = await comment.query_selector('span[class*="like-count"]').text_content()
            except Exception:
                likes = "0"
            comment_list.append({
                "username": username,
                "text": text,
                "likes": likes
            })

        with open("comments.json", "w", encoding="utf-8") as f:
            json.dump(comment_list, f, indent=4, ensure_ascii=False)

        print(f"Saved {len(comment_list)} comments to comments.json")
        await browser.close()

if __name__ == "__main__":
    video_url = "https://www.tiktok.com/@anuu__21/video/7415536407583755525"
    asyncio.run(scrape_tiktok_video(video_url))
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

class CommentScraper:
    def __init__(self):
        # Set up Chrome WebDriver (Ensure chromedriver is installed)
        self.driver = webdriver.Chrome()

    def load_page(self, url):
        """Load the webpage and wait for it to load."""
        self.driver.get(url)
        self.driver.maximize_window()
        time.sleep(5)  # Allow time for JavaScript content to load

    def scroll_down(self):
        """Scroll down the page to load more comments."""
        for _ in range(3):  # Scroll 3 times to ensure all comments are loaded
            self.driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(3)

    def extract_comments(self):
        """Extract comments from the webpage."""
        try:
            # Wait for the comment section to load
            comments_section = WebDriverWait(self.driver, 40).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "your_comment_selector"))
            )

            # Find all comments
            comments = self.driver.find_elements(By.CSS_SELECTOR, "your_comment_selector")

            # Extract text from each comment
            comment_texts = [comment.text for comment in comments if comment.text.strip()]

            # Print extracted comments
            for i, comment in enumerate(comment_texts, 1):
                print(f"Comment {i}: {comment}")

            return comment_texts

        except Exception as e:
            print("Error extracting comments:", e)
            return []

    def close(self):
        """Close the browser."""
        self.driver.quit()

# === RUN THE SCRAPER ===
if __name__ == "__main__":
    url = "https://www.tiktok.com/@anuu__21/photo/7415536407583755525"  # Replace with actual URL

    scraper = CommentScraper()
    scraper.load_page(url)

    # Scroll to load comments if needed
    scraper.scroll_down()

    # Extract comments
    comments = scraper.extract_comments()
    print("\nExtracted Comments:", comments)

    # Close the browser
    scraper.close()
```
